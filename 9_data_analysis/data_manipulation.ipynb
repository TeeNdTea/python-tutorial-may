```python
## Handling missing values
''' when it is True there is a missing value
we will use, df.isnull()
'''
```

```python
df.head(20).isnull()
```

```python
df.isnull().any(axis = 0)
# df.head(26).isnull().any(axis = 1)
```

```python
'''to check how many missing values'''
df.isnull().sum()
```

```python
## Fill missing values with numbers
df_filled = df.head(19).fillna('MISSING')
print(df_filled)
```

```python
# Fill missing values with the mean of the columns
df['Sales_FILL'] = df['Sales'].fillna(df['Sales'].mean())
df.head(19)
```

```python
# Rename a col
rename = df. rename(columns={'Date' : 'Sales Date','Value' : 'Val'})
rename.head(5)
```

```python
# Change data types
df['new_sale'] = df['Value'].fillna(df['Value'].mean()).astype(int)
df.head()
```

```python
# Multiply the values of Sales by 2
df['MU_Sales'] = df['new_sale'].apply(lambda x:x*2)
df.head()
```

```python
# Data Aggregating & Grouping
group_mean = df.groupby('Product')['Value'].mean()
group_mean
```

```python
df.groupby(['Product','Region'])['Value'].sum()
```

```python
group_agg = df.groupby('Region')['Value'].agg(['mean','sum','count'])
group_agg
```

```python
## Marging & Joining DataFrames
# Create sample dataFrames
df1 = pd.DataFrame({'Key' : ['A','B','C','D'], 'Value1': [1,2,3,4]})
df2 = pd.DataFrame({'Key' : ['B','F','C','H'], 'Value2': [5,6,7,8]})
```

```python
df1
```

```python
df2
```

```python
# Merging DataFrames
pd.merge(df1,df2,on='Key',how='outer')
```

```python
pd.merge(df1,df2,on='Key',how='inner')
```

```python
pd.merge(df1,df2,on='Key',how='left')
```

```python
pd.merge(df1,df2,on='Key',how='right')
```
