## Introduction to Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,0]

# Create a line plot
plt.plot(x,y)
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.title('Basic Line Plot')
```

```python
# Create customised line plot
x = [1,2,3,4,5]
y = [2,4,6,8,0]
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize=9)
plt.grid(True)
```

```python
# Multiple Plots
x = [1,2,3,9]
y = [1,3,4,5]
z = [6,7,8,0]

plt.figure(figsize=(7,5))
plt.subplot(2,2,1)      # rows cols side
plt.plot(x,y,color='purple')
plt.title('Plot-1')

plt.subplot(2,2,2)
plt.plot(y,z,color='green')
plt.title('Plot-2')

plt.subplot(2,2,3)
plt.plot(x,z,color='red')
plt.title('Plot-3')

plt.subplot(2,2,4)
plt.plot(z,x,color='blue')
plt.title('Plot-4')
```

```python
# Bar Plot
elems = ['A','B','C','D','E']
vals = [5,4,7,2,3]
plt.bar(elems,vals,color='orange')
```

### Histogram

```python
data = [1,2,2,3,3,3,4,5,5,6,8,9]
plt.hist(data,bins=5,color='yellow',edgecolor='black')
```

```python
# Scatter Plot
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.scatter(x,y,color='red',marker='x')
```

### Pie Charts

```python
labels = ['a','b','c','d']
sizes = [30,20,40,10]
colors = ['orange','green','lightblue','purple']
explode = (0.2,0,0,0) # 3d / move slice

plt.pie(sizes,explode = explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow='True')
```

### Sales Data Visualization

```python
import pandas as pd
```

```python
sales_data_df = pd.read_csv('sales_data.csv')
sales_data_df.head(5)
```

```python
sales_data_df.info()
```

```python
# Plot total sales by products
total = sales_data_df.groupby('Product Category')['Total Revenue'].sum()
print(total)
```

```python
total.plot(kind='bar',color='teal')
```

```python
# plot sales trend over time
trend = (sales_data_df.groupby('Date')['Total Revenue'].sum().reset_index())
plt.plot(trend['Date'],trend['Total Revenue'],color='pink')
```
