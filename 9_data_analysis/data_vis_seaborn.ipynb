## Data Visualization with Seaborn


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Basic plotting with Seaborn
tips = sns.load_dataset('tips')
tips
```

```python
# Create a scatter plot
sns.scatterplot(x='total_bill',y='tip',color='red', data=tips)
plt.title('Scatter plot : Total Bill Vs Tip')
plt.show()
```

```python
# Line Plot
sns.lineplot(x='size',y='total_bill',color='purple',data=tips)
plt.title('Line Plot : Size Vs Total bill ')
plt.show()
```

```python
# Categorical Variable
## Bar Plot
sns.barplot(x='day',y='total_bill',color='pink',data=tips)
plt.show()
```

```python
## Box Plot
sns.boxplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
## Violin Plot
sns.violinplot(x='day',y='total_bill',color='orange',data=tips)
plt.show()
```

```python
## Histogram
sns.histplot(tips['total_bill'],bins=10,color='green',kde=True)
```

```python
### KDE plot
sns.kdeplot(tips['total_bill'],color='green',fill=True)
```

```python
## Pair Plot
sns.pairplot(
    tips,
    plot_kws={'color': 'teal'},      # Off-diagonal scatter plots
    diag_kws={'color': 'teal'}       # Diagonal histograms
)
plt.show()
```

### Heatmap

```python
# correlation
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='crest')
```
